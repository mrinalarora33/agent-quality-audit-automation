These pipelines ensure that data is:

- validated before storage

- easily accessible for analytics queries

SQLite is a lightweight relational database system that allows us to simulate how real analytics teams store structured datasets before analysis.

## Loading CSV Tables

The normalized tables generated in Notebook 1 will now be loaded into memory.

Each CSV file corresponds to one relational table.

```python
customers = pd.read_csv("data/customers.csv")
orders = pd.read_csv("data/orders.csv")
agents = pd.read_csv("data/agents.csv")
interactions = pd.read_csv("data/interactions.csv")

logger.info("CSV files successfully loaded.")
```

## Schema Validation

Before loading data into the database, it is important to validate the structure of each dataset.

Schema validation ensures that:

- expected columns are present
- column names match the data model
- data ingestion errors are avoided

This step mimics data validation checks used in production pipelines.

```python
logger.info("Validating table schemas")

print("Customers Columns:", customers.columns.tolist())
print("Orders Columns:", orders.columns.tolist())
print("Agents Columns:", agents.columns.tolist())
print("Interactions Columns:", interactions.columns.tolist())
```

<hr>

## Creating SQLite Database

Databases are the foundation of analytics systems because they allow efficient storage, querying, and management of structured datasets.

In this step we create a SQLite database called:

`customer_support.db`

Before creating the database, we check whether it already exists to avoid accidental overwrites.

```python
db_name = "customer_support.db"

if os.path.exists(db_name):
    logger.info("Database already exists.")
else:
    logger.info("Creating new database.")
```

```python
# Create database connection

conn = sqlite3.connect(db_name)

logger.info("Connected to SQLite database.")
```

## Writing Tables to SQLite Database

Now that the connection is established, we write each dataset into the database as a relational table.

This step transforms the CSV datasets into queryable database tables.

```python
customers.to_sql("customers", conn, if_exists="replace", index=False)
logger.info("Customers table written to database")
```

```python
orders.to_sql("orders", conn, if_exists="replace", index=False)
logger.info("Orders table written to database")
```

```python
agents.to_sql("agents", conn, if_exists="replace", index=False)
logger.info("Agents table written to database")
```

```python
interactions.to_sql("interactions", conn, if_exists="replace", index=False)
logger.info("Interactions table written to database")
```

## Confirming Successful Ingestion

To verify that the ingestion process completed successfully, we query the database to list all available tables.

```python
query = "SELECT name FROM sqlite_master WHERE type='table';"

tables = pd.read_sql(query, conn)

tables
```
